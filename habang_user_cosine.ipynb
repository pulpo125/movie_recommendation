```python
ratings=pd.read_sql("select*from ratings",conn)
ratings.tail()
```

```python
pvt=ratings.pivot_table(index='user_id',columns='movie_id',values='rating').fillna(0)
pvt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim=cosine_similarity(pvt,pvt)
con_sim_df=pd.DataFrame(data=cos_sim)
con_sim_df
```

```python
similar_user_id = con_sim_df[5].sort_values(ascending=False).index[1]
```
