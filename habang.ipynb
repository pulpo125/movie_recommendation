```python
def get_top_idx(movielens, genre_cols, n=1):
    top_rating_idx = []
    for col in genre_cols:
        top_rating_idx += list(movielens.loc[movielens[col] == 1, 'movie_id'].value_counts()[:n].index)
    
    return list(set(top_rating_idx))

def get_top_movies(movies, top_rating_idx):    
    top_rating_movies = []
    for idx in top_rating_idx:
        top_rating_movies.append(movies.loc[movies['movie_id'] == idx, 'title'].values[0])
    return top_rating_movies
```
